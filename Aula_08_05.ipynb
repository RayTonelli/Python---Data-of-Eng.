```python
#Exercico 5
import requests
json=requests.get("https://jsonplaceholder.typicode.com/posts")
posts=json.json()

for post in posts:
  if post["userId"]==10:
    print("Título: ", post["title"])
    print("Post: ", post["body"])
    print()
```
